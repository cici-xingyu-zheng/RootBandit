```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10*np.pi, 40) # looks like okay-ish sampling frequency
```

```python
# wave = np.sin(x/2)
# slopes = 1/2*np.cos(x/2)
wave = np.sin(x) 
slopes = np.cos(x)
```

```python
slope_angles = np.arctan(slopes)
```

```python
for i, angle in enumerate(slope_angles):
    if angle > np.pi/2:
        slope_angles[i] = - (np.pi - angle)
```

### The reasoning is following:

- If we approximate the curvy root $y$ by a sine wave, $y = sin(x)$, 
- then the slope would be $y' = cos(x)$, and we can get the corresponding angle, $\theta$, by using **arctan function**, and find the angle in $[ - \frac{\pi}{2} , \frac{\pi}{2} ]$ range;
- then we can get change of angle by getting $\Delta \theta_t = \theta_t - \theta_{t-1} $

```python
fig, ax = plt.subplots(figsize = (15, 1.5))
ax.plot(x, wave, label = 'wave = sin(x)')
ax.plot(x, slopes, label = 'slopes  = cos(x)')
# ax.plot(x, changes, label = 'changes = - sin(x)')
ax.set_xticks([0,  2*np.pi, 4*np.pi, 6*np.pi, 8*np.pi, 10*np.pi])
ax.set_xticklabels([0,'$2\pi$', '$4\pi$', '$6\pi$','$8\pi$','$10\pi$'], fontsize = 12)
plt.legend(loc = 1, fontsize = 14)
plt.show()
```

```python
slope_angles_padded = slope_angles.tolist()
```

```python
slope_angles_padded.append(0)
```

```python
slope_angles*180/np.pi
```

```python
slope_angles.shape
```

```python
slope_angle_changes = np.array(slope_angles_padded[1:]) - np.array(slope_angles_padded[:-1])
```

```python
slope_angle_changes.shape
```

```python
fig, ax = plt.subplots(figsize = (4, 6))

ax.scatter(slope_angles[:-1], slope_angle_changes[:-1])
ax.set_xticks([-np.pi/2, 0, np.pi/2])
ax.set_xticklabels(['$-\pi/2$', 0, '$\pi/2$'])
ax.set_yticks([-np.pi/2, 0, np.pi/2])
ax.set_yticklabels(['$-\pi/2$', 0, '$\pi/2$'])

plt.show()
```

```python
plt.hist(slope_angles*180/np.pi, bins = 10)
plt.hist(slope_angle_changes[:-1]*180/np.pi, bins = 9)
plt.show()
```
