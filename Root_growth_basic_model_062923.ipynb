```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a0 = 2 # constant accleration 
a1 = 0 # constant speed

ts = np.linspace(0, 10, 200)
dt = 10/200
a_t = [a0]*2*20 + [a1]*8*20 # accelerate till t > 2
```

```python
plt.plot(ts, a_t, label = 'a(t)')
plt.legend()
plt.show()
```

```python
v_t = []
v_t.append(0)
for i in range(len(ts)-1):
    v_t.append(v_t[-1]+ dt*a_t[i])
```

```python
plt.plot(ts, v_t, label = 'v(t)', c = 'C1')
plt.legend()
plt.show()
```

```python
l_t = []
l_t.append(0)
for i in range(len(ts)-1):
    l_t.append(l_t[-1]+ dt*v_t[i])
```

```python
plt.plot(ts, l_t, label = 'length(t)', c = 'C2')
plt.legend()
plt.show()
```

```python
plt.plot(ts, a_t, label = 'a(t)')
plt.plot(ts, v_t, label = 'v(t)')
plt.plot(ts, l_t, label = 'length(t)')
plt.legend()
plt.show()
```
