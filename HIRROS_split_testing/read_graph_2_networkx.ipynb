```python
example_root = nx.json_graph.adjacency_graph(root_data)
```

```python
import numpy as np
```

```python
pos = np.array([example_root.nodes[node]['pos'] for node in list(example_root.nodes())])
```

```python
mins = np.min(pos, 0)
maxes = np.max(pos, 0)
```

```python
example_root.graph['x_min'] = mins[0]
example_root.graph['y_min'] = mins[1]

example_root.graph['x_max'] = maxes[0]
example_root.graph['y_max'] = maxes[1]

example_root.graph['ratio'] = (maxes[0] - mins[0]) / (maxes[1] - mins[1])
```

```python
import matplotlib.pyplot as plt

def quick_plot(G):
    '''
    quick plotting for spatial graphs with attr "pos" as a length 2 coordinate vector  
    '''
    node_positions = {}
    color_dict = {0:'C1',  1: 'C0'}
    node_color = []
    
    for node in G.nodes:
        node_positions[node] = [- G.nodes[node]['pos'][0], - G.nodes[node]['pos'][1]]
        node_color.append(color_dict[G.nodes[node]['root_deg']])
        
    fig, ax = plt.subplots(figsize=(2,2/G.graph['ratio']))
   
    nx.draw(G, pos=node_positions, node_size= 20, ax = ax, node_color = node_color) 
    
    plt.tight_layout()
    plt.show()   
    
    fig.savefig('example_root.pdf')    
```

```python
quick_plot(example_root)
```

```python
### graph block ###

# primary roots:
primary = left['root']['geometry']['polyline']['point']

for i, node in enumerate(primary):
    G_left.add_node((float(node['@coord_x']), float(node['@coord_y'])),
                    t = float(node['@coord_th']),
                    root_deg = 0,
                    root_id = left['root']['@ID'])
    if i < (len(primary)-1):
        u = (float(primary[i]['@coord_x']), float(primary[i]['@coord_y']))
        v =  (float(primary[i+1]['@coord_x']), float(primary[i+1]['@coord_y']))
        G_left.add_edge(u, v)
```
